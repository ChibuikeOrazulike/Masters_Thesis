```python
import pandas as pd
```

```python
df = pd.read_csv('GWS_no_soil.csv')
```

```python
#This code calculates the average storage for each year, computes the difference between consecutive years, 
#and then calculates the average of these differences to obtain the average annual change in groundwater storage. 
#The result is then added to the DataFrame as a new column called 'average_annual_change'.
```

```python
# Calculate the average annual change in groundwater storage
storage_columns = [col for col in df.columns if col not in ['lat', 'lon']]
years = list(set([col.split('-')[0] for col in storage_columns]))
years.sort()

annual_changes = []
```

```python
for year in years[:-1]:
    next_year = str(int(year) + 1)
    year_columns = [col for col in storage_columns if col.startswith(year)]
    next_year_columns = [col for col in storage_columns if col.startswith(next_year)]

    year_storage = df[year_columns].mean(axis=1)
    next_year_storage = df[next_year_columns].mean(axis=1)

    annual_change = next_year_storage - year_storage
    annual_changes.append(annual_change)

average_annual_change = pd.concat(annual_changes, axis=1).mean(axis=1)

# Add the average annual change to the DataFrame
df['average_annual_change'] = average_annual_change

df.to_csv('GWS_no_soil_AAC.csv', index=None)
```
